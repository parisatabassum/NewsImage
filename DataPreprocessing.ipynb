```python
import os
import requests  # to get image from the web
import shutil

f = open(os.path.join(r'C:\Users\yuxia\Documents\CS7311_Project\FIREWHEEL\data',"content2019-03-v3.tsv"), "r",encoding="utf-8")
next(f)
for line in f:
    image_url = line.split("\t")[3]
    image_id=line.split("\t")[4]
    filename = os.path.join('img_3',image_id+".jpg")
    r = requests.get(image_url, stream=True,headers={'User-agent': 'Mozilla/5.0'})
    if r.status_code == 200:
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
    else:
        print("img can't be loaded")
```

```python
f = open(os.path.join(r'C:\Users\yuxia\Documents\CS7311_Project\FIREWHEEL\data',"content2019-03-v3.tsv"), "r",encoding="utf-8")
next(f)
with open('train_03.csv', 'a',encoding="utf-8") as the_file:
    header="img_id, img_url, title"
    the_file.write(header+"\n")
    for line in f:
        image_url = line.split("\t")[3].split("/")[-1]
        image_id=line.split("\t")[4]+".jpg"
        title=line.split("\t")[6]
        the_file.write(image_id+","+image_url+","+title+"\n")
```

```python
from nltk import pos_tag, word_tokenize
f = open("train_01.csv", "r",encoding="utf-8")
next(f)
with open('title_01_tokens.csv', 'a',encoding="utf-8") as the_file:
    header="img_id,img_url,title,title_eng,name_token"
    the_file.write(header+"\n")
    for line in f:
        title_eng=line.split(",")[3]
        title_tags=pos_tag(word_tokenize(title_eng))
        title_names = [ x for x, y in title_tags if y  == 'NNP' ]
        title_names_string = ' '.join(title_names)
        the_file.write(line.rstrip("\n")+","+title_names_string.rstrip("\n")+"\n")
```
