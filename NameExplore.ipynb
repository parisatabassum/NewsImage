```python
import pandas as pd
dataframe=pd.read_csv("title_tokens_bk.csv")
names=dataframe['name_token'].to_list()
```

```python
cleanedNames = [x for x in names if str(x) != 'nan']
```

```python
print(cleanedNames)
```

```python
split_name_list=[]
for name in cleanedNames:
    split_name = name.split(' ')
    for token in split_name:
        if token != '"' and token != '-' :
            split_name_list.append(token)
```

```python
name_list=[]
for name in cleanedNames:
        if name != '"' and name != '-' :
            name_list.append(name)
```

```python
import pandas
from collections import Counter
def df_word_count(name_list):
    name_counts = Counter(name_list)
    appen_1 = []
    appen_2 = []
    for i in name_counts.keys():
        appen_1.append(i)
    for j in name_counts.values():
        appen_2.append(j)
    data = {"word":appen_1 , "count":appen_2}
    dfword = pd.DataFrame(data)
    dfword = dfword.sort_values(by='count', ascending=False)
    dfword = dfword.reset_index()[["word","count"]]
    return dfword
```

```python
dfword=df_word_count(name_list)
dfword=dfword.iloc[:20,:]
dfword
```
